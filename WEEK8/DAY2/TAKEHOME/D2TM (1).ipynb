# Building a model on features selected

```python
x = df1[['wt','yr','origin']]
y = df1.mpg
```

```python
import statsmodels.api as sm
x_constant = sm.add_constant(x)
LR = sm.OLS(y,x_constant).fit()
LR.summary()
```

# Transformation of 3 columns using SQRT

```python
df2 = df1[['wt','yr','origin']]
```

```python
df3 = df2.transform(lambda z: np.sqrt(z))
x = df3[['wt','yr','origin']]
y = np.sqrt(y)
x_constant = sm.add_constant(x)
LR = sm.OLS(y,x_constant).fit()
LR.summary()
```

# LOG transform

```python
df4 = df1[['wt','yr','origin']]
df5 = df4.transform(lambda z: np.log(z))
x = df5[['wt','yr','origin']]
y = np.sqrt(y)
x_constant = sm.add_constant(x)
LR = sm.OLS(y,x_constant).fit()
LR.summary()
```

# Inverse transform

```python
df6 = df1[['wt','yr','origin']]
```

```python
df7 = df6.transform(lambda z: (1/z))
x = df7[['wt','yr','origin']]
y = np.sqrt(y)
x_constant = sm.add_constant(x)
LR = sm.OLS(y,x_constant).fit()
LR.summary()
```

## Inference --- Log transform performed the best.

# Backward elimination : checking for p value

```python
x=df1.drop("mpg",axis=1)
y=df1.mpg
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
cols = list(x.columns)
pmax = 0
while(len(cols)>0):
    x = df1[cols]
    x_new = sm.add_constant(x)
    LR = sm.OLS(y,x_new).fit()
    p = pd.Series(LR.pvalues.values[1:],index=cols)
    pmax = max(p)
    feature_name = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_name)
    else:
        break
selected_feature = cols
selected_feature
```
