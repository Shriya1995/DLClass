# Confusion Matrix

```python
def create_conf_mat(test_class_set, predictions):
    if (len(test_class_set.shape) != len(predictions.shape) == 1):
        return print('Arrays entered are not 1-D.\nPlease enter the correctly sized sets.')
    elif (test_class_set.shape != predictions.shape):
        return print('Number of values inside the Arrays are not equal to each other.\nPlease make sure the array has the same number of instances.')
    else:
        # Set Metrics
        test_crosstb_comp = pd.crosstab(index = test_class_set,
                                        columns = predictions)
        # Changed for Future deprecation of as_matrix
        test_crosstb = test_crosstb_comp.values
        return test_crosstb
```

```python
conf_mat = create_conf_mat(y_test, y_pred)
sns.heatmap(conf_mat, annot=True, fmt='d', cbar=False)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Actual vs. Predicted Confusion Matrix')
plt.show()

```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
conf_matrix=pd.DataFrame(data=cm,columns=['Normal:0','Abnormal:1'],index=['Actual:0','Actual:1'])
plt.figure(figsize = (8,5))
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="YlGnBu")
```

```python
TN=cm[0,0]
TP=cm[1,1]
FN=cm[1,0]
FP=cm[0,1]
sensitivity=TP/float(TP+FN)
specificity=TN/float(TN+FP)
```

```python
print(sensitivity)
print(specificity)
```

# Backward Elimination

```python
def backward_elimination(x=None,y=None):
    features= list(x.columns) # init
    i = 0
    model = {}
    p_series = {}

    while(len(features)>0):
        x = x[features] # gets updated after each iter
        x['const'] = 1
        model[i] = sm.Logit(endog=y,exog=x).fit()
        p_series[i] = pd.Series(data=model[i].pvalues.values,index=x.columns)
        # drop feature with max p_value . other than constant
        fmax = p_series[i].drop('const').idxmax()
        pmax = p_series[i][fmax]

        if pmax>0.05:
            features.remove(fmax)
            i = i+1
        else:
            return model,p_series,i
```
