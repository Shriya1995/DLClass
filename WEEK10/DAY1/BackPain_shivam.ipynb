```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("backpain.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(data.columns[13],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.columns = ['Pelvic Incidence','Pelvic Tilt','Lumbar Lordosis Angle','Sacral Slope','Pelvic Radius', 
                'Spondylolisthesis Degree', 'Pelvic Slope', 'Direct Tilt', 'Thoracic Slope', 
                'Cervical Tilt','Sacrum Angle', 'Scoliosis Slope','Outcome']

corr = data.corr()

# Set up the matplot figure
f, ax = plt.subplots(figsize=(12,9))

#Draw the heatmap using seaborn
sns.heatmap(corr, cmap='inferno', annot=True)
```

```python
#Convert two class names to 0 and 1
data.Class_att = [1 if each == "Abnormal" else 0 for each in data.Class_att]
```

```python
data.describe()
```

```python
y = data.Class_att.values
df = data.drop(["Class_att"],axis=1)

x = (df-np.min(df))/(np.max(df)-np.min(df)).values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state =42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
