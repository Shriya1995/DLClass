# 4. Filling up NAN values wherever needed

```python
def impute_with_mean(col):
    col_data = col
    missing_data = col_data.isnull().sum()
    if missing_data > 0:
        col_mean = col_data.mean()
        col_data.fillna(col_mean, inplace=True)
        col = col_data
    return col
```

### Considering 'Release Clause' column

```python
df1['Release Clause'] = impute_with_mean(df1['Release Clause'])
print (df1['Release Clause'].head(2))
```

## Generating Tables for each Position

## General Function for Generating all the tables

```python
def generate(pos):
    df_i = df4.where(df4['Position']==pos)
    return df_i.dropna().head()
```

```python
# Trying to loop all ...
import os
pos = ['RF', 'ST', 'LW', 'GK', 'RCM', 'LF', 'RS', 'RCB', 'LCM', 'CB',
       'LDM', 'CAM', 'CDM', 'LS', 'LCB', 'RM', 'LAM', 'LM', 'LB', 'RDM',
       'RW', 'CM', 'RB', 'RAM', 'CF', 'RWB', 'LWB']
for i in pos:
    dfALL = generate(i)       
    list_of_datasets = [dfRF, dfST, dfLW, dfGK, dfRCM, dfLF, dfRS, dfRCB, dfLCM, dfCB,
       dfLDM, dfCAM, dfCDM, dfLS, dfLCB, dfRM, dfLAM, dfLM, dfLB, dfRDM,
       dfRW, dfCM, dfRB, dfRAM, dfCF, dfRWB, dfLWB]
    output_path = 'Tables'
    # in Python you can loop through collections of items directly, its pretty cool.
    # with enumerate(), you get the index and the item from the sequence, each step through
    for index, dataset in enumerate(list_of_datasets):
        dataset = dfALL
    # Export to CSV
        filepath = os.path.join(output_path, 'dataset_'+str(index)+'.csv')
        dataset.to_csv(filepath,index=False)
        print(dataset.head())
```
