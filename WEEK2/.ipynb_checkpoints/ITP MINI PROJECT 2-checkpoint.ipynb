```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns",None)
df1=pd.read_csv("fifa.csv")
df1.head()
```

```python
#check if any empty rows:
#df1.isnull().sum()
```

```python
#Dropping unwanted columns since these columns don't help in prediction
df1 = df1.drop(['Photo','Flag','Club Logo','Nationality','Loaned From','Club'],axis=1)
df1.head()
```

```python
df1.Value.head()
```

```python
#Removing euro symbol and M from value column and converting into float
df1['Value'] = df1['Value'].map(lambda x: str(x)[:-1])
df1['Value'] = df1['Value'].map(lambda x: str(x)[1:])
df1['Value'].head()

                

```

```python
#Removing euro symbol and M from wage column
df1["Wage"]=df1["Wage"].map(lambda x:str(x)[:-1])
df1["Wage"]=df1["Wage"].map(lambda x:str(x)[1:])
df1["Wage"].head()
```

```python
#Working with joined column
df1=df1[pd.notnull(df1["Joined"])]
df1['Joined'].isnull().sum()


```

```python
#extact year from joined
df1['Joined'] = pd.to_datetime(df1['Joined'])
df1['Joined'] = df1['Joined'].dt.year
df1['Joined'] = df1['Joined'].astype(np.int64)
df1['Joined'].head(2)
```

```python
#Release clause
df1["Release Clause"]=df1["Release Clause"].map(lambda x:str(x)[:-1])
df1["Release Clause"]=df1["Release Clause"].map(lambda x:str(x)[1:])
df1["Release Clause"].head()
```

```python
#Remove the suffix lbs from weight
df1["Weight"]=df1["Weight"].map(lambda x: str(x)[:-3])
df1["Weight"]
```

```python
#Removing quotation mark in height
def cleanheight(height):
    if (type(height) is not float):
        height=height.replace("'",".")
    return float(height)

df1["Height"]=df1["Height"].apply(cleanheight)

df1["Height"]
        
```

```python
#Check for missing values and do a mean imputation where necessary
df1["Overall"].fillna(df1["Overall"].mean(),inplace=True)
df1["Value"].fillna(df1["Value"].mean(),inplace=True)
df1["Wage"].fillna(df1["Wage"].mean(),inplace=True)
df1["Height"].fillna(df1["Height"].mean(),inplace=True)
df1["Weight"].fillna(df1["Weight"].mean(),inplace=True)
df1["Release Clause"].fillna(df1["Release Clause"].mean(),inplace=True)
df1['International Reputation'].fillna(df1['International Reputation'].mean(),inplace=True)

```

```python
#Exploratory analysis
sns.distplot(df1["Overall"])
```

```python
import seaborn as sns
sns.set()
cols=["Overall","Value","Wage","International Reputation","Height","Weight","Release Clause"]


sns.pairplot(df1[cols],height=2.5)
plt.show();
```

```python
#Generate a table containing the top 20 players ranked by Overall score and whose contract expires in 2020.
result=df1.loc[df1["Contract Valid Until"].dt.year==pd.to_datetime('01-01-2020').year] 

result=df1.sort_values("Overall",ascending=False).head(20)
result.head()
```

```python
#Average wage of this set of players:
result["Wage"].mean(skipna=True)
```

```python
#Average age
result["Age"].mean(skipna=True)
```
