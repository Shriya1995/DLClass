```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_columns",None)
df=pd.read_csv("fifa.csv")
df.head()
```

```python
#check if null value
#df.isnull().sum()
```

```python
#Dropping unwanted columns
pd.set_option("display.max_columns",None)
df1=df.drop(['Jersey Number','Photo','Flag','Club Logo','Jersey Number','Loaned From','Body Type'],axis=1)
df1.head(2)
```

```python
df1.shape  #after dropping
```

```python
df.shape #before dropping
```

```python
#Value problem
df1["Value"]=df1["Value"].map(lambda x:str(x)[:-1])
df1["Value"]=df1["Value"].map(lambda x:str(x)[1:])
df1["Value"].head()
```

```python
#Converting the values to float type
df1["Value"]=pd.to_numeric(df1.Value)
df1.head(2)
```

```python
#Removing Euro Symbol and K from Wage Column
df1["Wage"]=df1["Wage"].map(lambda x:str(x)[:-1])
df1["Wage"]=df1["Wage"].map(lambda x:str(x)[1:])
df1.head(2)
```

```python
#Convert wage values into float
df1["Wage"]=pd.to_numeric(df1.Wage)
df1.head(2)
```

```python
#Working with Joined column:
df1=df1[pd.notnull(df1["Joined"])]
```

```python
df1["Joined"].isnull().sum()
```

```python
df1["Joined"]=pd.to_datetime(df1["Joined"])
df1["Joined"]=df1["Joined"].dt.year
df1["Joined"].head(2)
```

```python
df1['Joined'] = df1['Joined'].astype(np.int64)
```

```python
df1.loc["Joined"]=df1["Joined"]
```

```python
i = df1[((df1['Joined'] == '1970'))].index

df1['Joined'].drop(i)
```

```python
# Working with Contract valid until
```

```python
df1=df1[pd.notnull(df1["Contract Valid Until"])]
df1["Contract Valid Until"].isnull().sum()
```

```python
df1["Contract Valid Until"]=pd.to_datetime(df1["Contract Valid Until"])
df1["Contract Valid Until"]=df1["Contract Valid Until"].dt.year
```

```python
df1["Contract Valid Until"].head()
```

```python
#Height
df1=df1[pd.notnull(df1["Height"])]
df1["Height"].isnull().sum()
```

```python
def cleanheight():
    if (type(height) is not float):
        height=height.replace("'",".")
    return float(height) 



df1["Height"]=df1["Height"].apply(cleanheight)

df1["Height"].head()
```

```python
#Weight column
df1['Weight'] = df1['Weight'].map(lambda x : str(x)[:-3])
df1["Weight"]
```

```python
def x(col):
    col_data=col
    missing_data=col_data.isnull().sum()
    if missing_data>0:
        col_mean=col_data.mean()
        col_data.fillna(col_mean,inplace=True)
        col=col_data
    return(col)

    
```

```python
#consider release clause column:
df1["Release Clause"]=x(df1["Release Clause"])
print(df1["Release Clause"].head())
```
