**3 Blood group of patients of the hospital is given below:**

| Blood group | Count |
| ------ | --- |
| O | 44%|
| A | 45% |
| B | 8% |
| AB | 3% |


Observed frequency

| Blood group | Count |
| ----- | ----|
| O | 67 |
| A | 83 |
| B | 29 |
| AB | 8 |

At 5% level of significance do you have evidence to conclude that there are differences in the distribution of blood groups among patients that is expected as per our prior knowledge.



# Here the Null Hyp is that the prop of O = prop of A = prop of B = prop of AB

```python
import scipy.stats as stats
import scipy

observed_values=scipy.array([67, 83, 29, 8])
n=observed_values.sum()
expected_val = scipy.array([n*0.44, n*0.45, n*0.08, n*0.03])
from scipy.stats import chisquare
chisquare(observed_values,expected_val)
```

# Clearly the null hyp is rejected here as the p value is < 0.05 and hence there is strong evidence that there are differences in the distribution of blood groups.
