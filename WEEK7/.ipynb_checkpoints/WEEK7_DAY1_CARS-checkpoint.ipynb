# Deep copy: in this, changes are not reflected.

```python
import copy
B=copy.deepcopy(df)
```

```python
B
```

```python
B["Price"]=np.log(df["Price"])
```

```python
B["Price"].plot(kind="box")
```

```python
B["Price"].skew()
```
